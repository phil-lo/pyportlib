```python
import pandas as pd
import pyportlib
pyportlib.set_client_dir()
update_data = False
```

# Portfolio Summary Example

```python
ptf = pyportlib.Portfolio(account='questrade_tfsa', currency="CAD")
benchmark = pyportlib.Portfolio(account='bench_tfsa', currency='CAD')

q = pyportlib.QuestradeConnection(account_name='tfsa')
```

```python
if update_data:
    ptf.update_data(fundamentals_and_dividends=False)
    benchmark.update_data(fundamentals_and_dividends=False)
```

```python
if update_data:
    q.update_transactions(ptf)
```

```python
cash_cad = q.get_balances().get("combinedBalances")[0].get("cash")
equity_cad = q.get_balances().get("combinedBalances")[0].get("totalEquity")
mv_cad = q.get_balances().get("combinedBalances")[0].get("marketValue")

print("Questrade")
print(f"Cash CAD: ${round(cash_cad, 2)}")
print(f"Market Value CAD: ${round(mv_cad, 2)}")
print(f"Total Equity CAD: ${round(equity_cad, 2)}\n")

print("pyportlib")
print(f"Cash CAD: ${round(ptf.cash_history.iloc[-1], 2)}")
print(f"Market Value CAD: ${round(ptf.market_values.iloc[-1], 2)}")
print(f"Total Equity CAD: ${round(ptf.market_values.iloc[-1], 2) + round(ptf.cash_history.iloc[-1], 2)}")
```

## Daily Return By Position

```python
date = pyportlib.dates_utils.last_bday().replace(hour=0, minute=0, second=0, microsecond=0)
dol = ptf.daily_total_pnl(end_date=date).loc[:,ptf.open_positions(date).keys()].T
pct = pd.DataFrame.from_dict({k: round(pos.prices.pct_change().loc[date]*100, 2) for k, pos in ptf.open_positions(date).items()}, orient="index", columns=['Daily %'])
dol.columns = ["Daily $"]
pnl = pd.concat([dol, pct], axis=1)
pnl
```

## Total Daily Return

```python
dol_tot = ptf.daily_total_pnl(end_date=date).loc[:,ptf.open_positions(date).keys()].sum(axis=1).iloc[0]
pct_tot = ptf.pct_daily_total_pnl(end_date=date, include_cash=True).item()

print(f"Date: {date.date()}")
print(f"Daily $: {round(dol_tot, 2)}$")
print(f"Daily %: {round(pct_tot*100, 2)}%")
```

## Stats
### Portfolio

```python
lookback = '1y'
quantile = 0.975
print(f"{lookback} skewness: {pyportlib.stats.skew(ptf, lookback=lookback, include_cash=True)}")
print(f"{lookback} kurtosis: {pyportlib.stats.kurtosis(ptf, lookback=lookback, include_cash=True)}")
print(f"{lookback} vol: {pyportlib.stats.annualized_volatility(ptf, lookback=lookback, include_cash=True)}")
print(f"{lookback} beta: {pyportlib.stats.beta(ptf, benchmark=benchmark, lookback=lookback, include_cash=True)}")
print(f"{lookback} alpha: {pyportlib.stats.alpha(ptf, benchmark=benchmark, lookback=lookback, include_cash=True)}")
print(f"{lookback} VaR {quantile}: {pyportlib.stats.value_at_risk(ptf, '1y', quantile=quantile, include_cash=True)}")
```

### Benchmark

```python
lookback = '1y'
quantile = 0.975
print(f"{lookback} skewness: {pyportlib.stats.skew(benchmark, lookback=lookback)}")
print(f"{lookback} kurtosis: {pyportlib.stats.kurtosis(benchmark, lookback=lookback)}")
print(f"{lookback} vol: {pyportlib.stats.annualized_volatility(benchmark, lookback=lookback)}")
print(f"{lookback} VaR {quantile}: {pyportlib.stats.value_at_risk(benchmark, '1y', quantile=quantile)}")
```

```python
lk = "5y"
pyportlib.plots.rolling_var(ptf, lookback=lk, rolling_period=252, quantile=0.95, figsize=figsize, include_cash=True)
roll_var = pyportlib.stats.rolling_var(ptf, lookback=lk, rolling_period=252, quantile=0.95, include_cash=True)
roll_var = roll_var * ptf.market_values.loc[roll_var.index]

roll_var.plot(figsize=figsize, title="Portfolio VaR 95% $")
```

```python
mkt_val = ptf.market_values.loc[date]
lk = "1y"
var99 = round(pyportlib.stats.value_at_risk(ptf, lookback=lk, quantile=0.99, include_cash=True), 3)
var975 = round(pyportlib.stats.value_at_risk(ptf, lookback=lk, quantile=0.975, include_cash=True), 3)
var95 = round(pyportlib.stats.value_at_risk(ptf, lookback=lk, quantile=0.95, include_cash=True), 3)

print(f"VaR 99%: {var99} %")
print(f"VaR 97.5%: {var975} %")
print(f"VaR 95%: {var95} %\n")

print(f"VaR 99%: {round(var99 * mkt_val, 2)}$")
print(f"VaR 97.5%: {round(var975 * mkt_val, 2)}$")
print(f"VaR 95%: {round(var95 * mkt_val, 2)}$")
```
