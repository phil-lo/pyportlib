```python
import pyportlib
pyportlib.set_client_dir()
sp500 = pyportlib.SP500().constituents
```

```python
universe = {ticker: pyportlib.Position(ticker, 'USD') for ticker  in sp500}
```

```python
pyportlib.plots.returns(universe.get('MSFT'), benchmark=universe.get('AAPL'))
```

```python
pyportlib.plots.rolling_vol(universe.get('MSFT'), benchmark=universe.get('AAPL'), figsize=(10,6), lookback='3y')
```

```python
pyportlib.plots.rolling_beta(universe.get('MSFT'), lookback='5y', benchmark=universe.get('AAPL'), figsize=(10,5))
```

```python
pyportlib.plots.rolling_sharpe(universe.get('MSFT'), lookback='5y', benchmark=universe.get('AAPL'), figsize=(10,5))
```

```python
pyportlib.plots.distribution(universe.get('MSFT'), lookback='5y', figsize=(10,5), resample='W')
```
